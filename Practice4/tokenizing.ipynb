```python
import nltk
import nltk.tokenize
```

```python
text = open("c:\\Users\\maria\\Downloads\\metamorphosis.txt").read()
text
```

```python
whitespace = nltk.tokenize.WhitespaceTokenizer()
print(whitespace.tokenize(text))
#WhitespaceTokenizer - простой (и в реализации, и для понимания) токенайзер, выполняющий ту же функцию, что и split(). Однако простота этого токенайзера может служить и недостатком в случаях работы с сокращениями, аббревиатурами или знаками препинания внутри слов.
```

```python
spacetoken = nltk.tokenize.SpaceTokenizer()
print(spacetoken.tokenize(text))
#SpeceTokenizer своей простотой напоминает WhitespaceTokenizer. По функционалу совпадает с методом s.split(‘ ‘). В недостатки можно отнести, пожалуй, только то, что он обрабатывает последовательные пробелы как единый разделитель.
```

```python
tok = nltk.tokenize.SpaceTokenizer()
tokenized_text = tok.tokenize(text)
print(tokenized_text)
mwe = nltk.tokenize.MWETokenizer(separator=' ')
mwe.add_mwe(('Gregor', 'Samsa'))
print(mwe.tokenize(tokenized_text))
#MWETokenizer создан для обработки многословных выражений (multi-word expressions). Это полезно при работе с устойчивыми выражениями. К недостаткам этого токенайзера можно отнести факт того, что его эффективность зависит от полноты и точности используемого им списка выражений. 
```

```python
toktok = nltk.tokenize.ToktokTokenizer()
print(toktok.tokenize(text))
#ToktokTokenizer - простой токенайзер. Он основан на последовательном применении к исходному тексту списка регулярных выражений. К минусам можно отнести то, что нужно проверять и дополнительно очищать текст от лишних знаков препинания и иных символов после применения этого токенайзера.
```

```python
treeword = nltk.tokenize.TreebankWordTokenizer()
print(treeword.tokenize(text))
#TreebankWordTokenizer отлично обрабатывает знаки препинания и сокращения в тексте, так как он рассматривает знаки препинания как отдельные лексемы, что позволяет лучше анализировать структуру предложения. Он также правильно обрабатывает сокращения, сохраняя целостность таких слов, как "o'clock". Недостатком является то, что TreebankWordTokenizer имеет тенденцию к чрезмерному разделению определенных токенов, особенно в случаях слов, написанных через дефис, или аббревиатур, которые разделены на отдельные части. Это может привести к потенциальным проблемам при попытке восстановить исходные слова или при работе с лексикой, относящейся к конкретной области.
```

```python
word_tok = nltk.tokenize.NLTKWordTokenizer()
print(word_tok.tokenize(text))
#NLTKWordTokenizer в какой-то мере можно считать улучшенной версией TreebankWordTokenizer, так как он имеет несколько адаптированных списков сокращений.
```

```python
word_punct = nltk.tokenize.WordPunctTokenizer()
print(word_punct.tokenize(text))
#WordPunctTokenizer отличается от WhitespaceTokenizer тем, что отделяет знаки препинания и иные символы пунктуации от слова, и возвращает их в качестве отдельных токенов.
```

```python
tweet = nltk.tokenize.TweetTokenizer()
print(tweet.tokenize(text))
#TweetTokenizer это токенайзер, который поддерживает специфику коротких сообщений Twitter. 
```

```python
text_tiling = nltk.tokenize.TextTilingTokenizer()
print(text_tiling.tokenize(text))
#TextTilingTokenizer специально разработан для разделения текста на тематические разделы.  Одним из недостатков является то, что TextTilingTokenizer может работать оптимально не для всех типов текста или языков.  
```

```python
tab = nltk.tokenize.TabTokenizer()
print(tab.tokenize(text))
#TabTokenizer - брат-близнец SpaceTokenizer, только для разделения строки использует символ табуляции, что совпадает с применением метода s.split(‘\t’).
```

```python
syllable = nltk.tokenize.SyllableTokenizer()
print(syllable.tokenize(text))
#SyllableTokenize разбивает слова на слоги. Он может быть весьма удобен при работе с языками, имеющими сложную структуру слогов. Своим функционалом он облегчает их анализ и обработку. Однако он справляется далеко не со всеми языками и их структурами слогов.
```

```python
xpr = nltk.tokenize.SExprTokenizer()
print(xpr.tokenize(text))
#SExprTokenizer помогает токенизировать символические выражения. Он может обрабатывать сложные вложенные структуры, что может быть весьма удобно для определенных задач. Одним из недостатков является то, что он может быть не самый эффективный вариантом для простых задач токенизации. 
```

```python
xp = nltk.tokenize.RegexpTokenizer(pattern = '[A-Z]\w+')
print(xp.tokenize(text))
#RegexpTokenizer позволяет токенизировать текст на основе регулярных выражений. Это помогает в извлечении значимых фрагментов текста. Одним из недостатков является то, что может быть сложно придумать идеальное регулярное выражение, которое фиксирует все желаемые токены без каких-либо ошибок. 
```

```python
punkt = nltk.tokenize.PunktSentenceTokenizer()
print(punkt.tokenize(text))
#PunktSentenceTokenizer специально разработан для токенизации предложений. Он использует предварительно обученные модели для точного определения границ предложений. Одним из недостатков является то, что PunktSentenceTokenizer может работать не идеально для всех языков. Он основан на предварительно обученных моделях, которые могут быть не оптимизированы для конкретных случаев использования. Это также может привести к проблемам с неоднозначными границами предложений или необычными текстовыми форматами.
```

```python
line = nltk.tokenize.LineTokenizer()
print(line.tokenize(text))
#LineTokenizer идентичен SpaceTokenizer, только для разделения строки использует символ новой строки. Одним из недостатков является то, что если текст не соответствует четкой линейной структуре, то этот токенайзер может не подойти. 
```

```python

legality_principle = nltk.LegalitySyllableTokenizer(text)
print(legality_principle.tokenize(text))
#LegalitySyllableTokenizer разбивает слова в тексте по слогам. Oдним из недостатков является то, что он может подходить не для всех языков, так как структуры слогов могут значительно различаться в разных языках. Даже в моем тексте на английском он ингда странно разбивает слова.
```

```python
blank_line = nltk.tokenize.BlanklineTokenizer()
print(blank_line.tokenize(text))
#BlanklineTokenizer разбивает текст, рассматривая в качестве разделителя последовательность пустых строк, содержащих символы пробела или табуляции, то есть если текст не имеет четких границ с пустыми строками, этот токенайзер не даст значимых результатов.
```

```python
segmenter = nltk.tokenize.StanfordSegmenter()
print(segmenter.tokenize(text))
#StanfordSegmenter в теории классно справляется с сегментацией слов некоторых языков. Однако чтобы его запустить нужно попотеть, так как он опирается на внешние библиотеки и ресурсы, что означает, что потребуется обеспечить совместимость и предоставить необходимые файлы данных.
```

```python
def calculate_lexical_diversity(text):
    words = text.split()
    unique_words = set(words)
    total_words = len(words)
    lexical_diversity = len(unique_words) / total_words
    print(lexical_diversity)
```

```python
lexical_diversity = calculate_lexical_diversity(text)
```

```python
class MyTokenizer:
    def __init__(self, text):
        self.text = text

    def tokenize(self):
        tokens = self.text.split()
        return tokens
```

```python
tokenizer = MyTokenizer(text)
tokens = tokenizer.tokenize()
print(tokens)
```
