```python
import nltk
import nltk.tokenize
```

```python
tok = nltk.tokenize.SpaceTokenizer()
tokenized_text = tok.tokenize(text)
print(tokenized_text)
mwe = nltk.tokenize.MWETokenizer(separator=' ')
mwe.add_mwe(('Gregor', 'Samsa'))
print(mwe.tokenize(tokenized_text))
#MWETokenizer      (multi-word expressions).       .        ,            . 
```

```python
def calculate_lexical_diversity(text):
    words = text.split()
    unique_words = set(words)
    total_words = len(words)
    lexical_diversity = len(unique_words) / total_words
    print(lexical_diversity)
```

```python
lexical_diversity = calculate_lexical_diversity(text)
```

```python
class MyTokenizer:
    def __init__(self, text):
        self.text = text

    def tokenize(self):
        tokens = self.text.split()
        return tokens
```

```python
tokenizer = MyTokenizer(text)
tokens = tokenizer.tokenize()
print(tokens)
```
